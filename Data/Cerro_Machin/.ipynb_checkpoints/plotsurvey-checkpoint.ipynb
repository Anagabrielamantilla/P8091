```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("Grav_anomaliaresidual.csv", sep=";")
df2 = pd.read_csv("Mag_TMIresidual.csv", sep=";")

plt.figure()
plt.scatter(df1["X"], df1["Y"], label="Gravimetría", color="blue")
plt.scatter(df2["X"], df2["Y"], label="Magnetometría", color="red", s=15)

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.savefig("plotsurvey.png", dpi=300, format="png")
plt.show()
```
