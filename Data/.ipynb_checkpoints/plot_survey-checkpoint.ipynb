```python
import numpy as np
import matplotlib.pyplot as plt

# Paths (from this workspace)
grav_path = 'Datos_Grav_SGC.npy'
mag_path  = 'Datos_Mag_SGC.npy'

# Load arrays: columns assumed [X, Y, Z, value]
grav = np.load(grav_path)
mag  = np.load(mag_path)

xg, yg = grav[:, 0], grav[:, 1]
xm, ym = mag[:, 0],  mag[:, 1]

print('Grav shape:', grav.shape)
print('Mag  shape:', mag.shape)
print('Grav X range:', (xg.min(), xg.max()))
print('Grav Y range:', (yg.min(), yg.max()))
print('Mag  X range:', (xm.min(), xm.max()))
print('Mag  Y range:', (ym.min(), ym.max()))
```

```python
# Overlay plot (X vs Y)
fig, ax = plt.subplots(figsize=(8, 7))

# Two scatters so matplotlib assigns different default colors automatically
ax.scatter(xg, yg, s=50, label='Gravimetry')
ax.scatter(xm, ym, s=18, label='Magnetometry')

ax.set_title('Sensor locations overlay (X vs Y)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_aspect('equal', adjustable='box')
ax.legend()
ax.grid(True, linewidth=0.5)
plt.tight_layout()
plt.savefig("plot_survey.png", dpi=300, format="png")
plt.show()
```
